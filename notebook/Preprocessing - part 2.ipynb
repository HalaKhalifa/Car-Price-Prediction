- The data currently have an 'object' data type as we said before and need to be converted to either 'int' or 'float' for numeric operations.
- The data contains Arabic numerals ('١٨٣٠٠٠'), and these need to be translated into their numeric equivalents. 
- The data contains Arabic words ('الف'), which should be converted to '000' or the numbers should be multiplied by 1000.


```python
df['Mileage']
```

```python
def clean_mileage(mileage):
    try:
        # Remove spaces, translate Arabic numerals to English
        translation_dict = {'٠': '0', '١': '1', '٢': '2', '٣': '3', '٤': '4', '٥': '5', '٦': '6', '٧': '7', '٨': '8', '٩': '9'}
        mileage = ''.join(translation_dict[char] if char in translation_dict else char for char in mileage.replace(' ', ''))

        # Remove Arabic words and commas
        mileage = mileage.replace('الف', '000').replace(',', '')

        # Convert to integer
        mileage = int(mileage)

        return mileage
    except:
        return None
```

```python
df['Mileage']= df['Mileage'].apply(clean_mileage)
```

#### Passenger Count

```python
#df['Passenger Count'].nunique()
df['Passenger Count'].unique()
```

```python
df['Passenger Count'].replace({'اكثر من 10': np.nan, 'مثال: 4 \+ 1': np.nan, '000': np.nan}, inplace=True, regex=True)
df['Passenger Count'] = df['Passenger Count'].str.replace('±', '+')
df['Passenger Count'].replace({'6x1': '6+1', '7.1': '7+1', '7 =1': '7+1', '7:1': '7+1', '4\\\\١': '4+1','2/1':'2'}, inplace=True, regex=True)
df['Passenger Count'] = df['Passenger Count'].replace('[^\d+]', '', regex=True)
arabic_numerals = {'١': '1', '٤': '4', '٥': '5', '٦': '6', '٧': '7', '٨': '8', '٩': '9' , '٠':'0'}
for arabic, numeral in arabic_numerals.items():
    df['Passenger Count'] = df['Passenger Count'].str.replace(arabic, numeral)
```
