- The data currently have an 'object' data type as we said before and need to be converted to either 'int' or 'float' for numeric operations.
- The data contains Arabic numerals ('ูกูจูฃููู'), and these need to be translated into their numeric equivalents. 
- The data contains Arabic words ('ุงูู'), which should be converted to '000' or the numbers should be multiplied by 1000.


```python
df['Mileage']
```

```python
def clean_mileage(mileage):
    try:
        # Remove spaces, translate Arabic numerals to English
        translation_dict = {'ู': '0', 'ูก': '1', 'ูข': '2', 'ูฃ': '3', 'ูค': '4', 'ูฅ': '5', 'ูฆ': '6', 'ูง': '7', 'ูจ': '8', 'ูฉ': '9'}
        mileage = ''.join(translation_dict[char] if char in translation_dict else char for char in mileage.replace(' ', ''))

        # Remove Arabic words and commas
        mileage = mileage.replace('ุงูู', '000').replace(',', '')

        # Convert to integer
        mileage = int(mileage)

        return mileage
    except:
        return None
```

```python
df['Mileage']= df['Mileage'].apply(clean_mileage)
```
